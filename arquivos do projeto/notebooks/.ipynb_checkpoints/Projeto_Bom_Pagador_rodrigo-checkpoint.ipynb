```python
base_final = base[base['Situacao']==1]
```

## Forma 1 de Balancear a Base
Utilizar amostragem aleatória para retirar a mesma quantidade de linhas que temos na situação 1

```python
np.random.seed(5)
amostra = np.random.choice(a = [0,1], size = 16204, replace = True, p =[0.3,0.7])
print(len(amostra[amostra ==1]))
print(len(amostra[amostra ==0]))
```

```python
base_0 = base[base['Situacao']==0]
base_0['Amostra'] = amostra
base_0
```

```python
base_final_forma1 = base_final.append(base_0[base_0['Amostra']==0].drop(columns=['Amostra'])) # Balanceando a Base
base_final_forma1
```

## Forma 2 de Balancear a Base
Utilizar a mesma quantidade de linhas que temos na situação 1

```python
base_final_forma2 = base_final.append(base[base['Situacao']==0].iloc[0:4796]) # Balanceando a Base
base_final_forma2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC, LinearSVC
from scipy import stats
from xgboost import XGBClassifier
from sklearn.ensemble import ExtraTreesClassifier
```

```python
previsores = base_final_forma2.iloc[:,[1,3,5,6]].values
previsores
```

```python
classe = base_final_forma2.iloc[:,7].values.reshape(-1, 1)
classe
```

```python
# Definindo base de treinamento e de teste
seed = 0
X_train, X_validation, y_train, y_validation = train_test_split(previsores, classe, test_size = 0.3, random_state = seed)
# random_state define a semente para a aleatóriedade
```

```python
# modelos que serão treinados
scoring = 'f1'

models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier(n_neighbors = 3)))
models.append(('DTC', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
models.append(('SVML', LinearSVC()))
models.append(('RF', RandomForestClassifier(n_estimators = 500)))
models.append(('XG', XGBClassifier()))
models.append(('ETC', ExtraTreesClassifier()))

models.append

results = []
names = []
msg_ = []

for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    msg_.append(msg)
    print(msg)
```

```python
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
msg_
```

```python
dtc = GaussianNB()
dtc.fit(X_train, y_train)
predictions = dtc.predict(X_validation)
print('Acurácia')
print(accuracy_score(y_validation, predictions))
print('\nMatrix de Confusão\n')
print(confusion_matrix(y_validation, predictions))
print('\nReport de Classificação\n')
print(classification_report(y_validation, predictions))
```
